1. Write a Python Program to Find LCM?

```python
class my_maths:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        
    def lcm(self):
        if self.x > self.y:
            maximum = self.x
        else:
            maximum = self.y
        while(True):
            if maximum % self.x == 0 and maximum % self.y == 0:
                lcm = maximum
                break
            maximum += 1
        return lcm
                
    def hcf(self):
        if self.x > self.y:
            minimum = self.y
        else:
            minimum = self.x
        for i in range(1, minimum + 1):
            if self.x % i == 0 and self.y % i == 0:
                hcf = i
        return hcf
    
    def addition(self):
        sum = self.x + self.y
        return sum
    
    def substraction(self):
        sub = self.x - self.y
        return sub
    
    def multiplication(self):
        mul = self.x*self.y
        return mul

    def division(self):
        div = self.x/self.y
        return div
    
    def d2b(self,n):
        binary_value = bin(n)
        return binary_value
    
    def d2o(self,n):
        octal_value = oct(n)
        return octal_value
    
    def d2h(self,n):
        hex_value = hex(n)
        return hex_value
    
    def ascii(self,value):
        ascii_value = ord(value)
        return ascii_value
```

```python
o = my_maths(54,24)
```

```python
o.lcm()
```
