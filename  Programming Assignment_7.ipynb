1. 1. Write a Python Program to find sum of array? 

```python
import logging
logging.basicConfig(filename = "array.log", level = logging.DEBUG, format = '%(asctime)s %(levelname)s %(message)s')

class array_operations():
    """
        This class is used for array operations.
        
        Written By: Vishwaraj Kolekar
        Version: 1.0
        Revisions: None
        
    """
    def __init__(self,A):
        self.A = A
        
    def sum_of_array(self):
        """
                Method Name: sum of array
                Description: This method adds all elements from list.
                Output: Gives sum of all elements.
                On Failure: Raise Exception

                Written By: Vishwaraj Kolekar
                Version: 1.0
                Revisions: None

        """
        logging.info("this is my sum of array function, my array is: %s", self.A)
        try:
            sum = 0
            for i in self.A:
                sum = sum + i
            return sum
            logging.info("executed successfully")
            
        except Exception as e :
            print("this is my error msg " , e )
            logging.error("Error has happended")
            logging.exception("Exception occured" + str(e))

    def large(self):
        """
            Method Name: large
            Description: This method return's largest element from the array.
            Output: Largest element in array.
            On Failure: Raise Exception

            Written By: Vishwaraj Kolekar
            Version: 1.0
            Revisions: None
        """
        logging.info("this is my find largest number function, my array is: %s", self.A)
        try:
            largest_number = 0
            for i in self.A:
                if i > largest_number:
                    largest_number = i
            return largest_number
            logging.info("executed successfully")
            
        except Exception as e :
            print("this is my error msg " , e )
            logging.error("Error has happended")
            logging.exception("Exception occured" + str(e))
        

    def rotateArray(self,d):
        """
            Method Name: rotateArray
            Description: This function rotate the elements of an array towards the left by the specified number of times. 
                In the left rotation, each element of the array will be shifted to its left by one position and the first 
                element of the array will be added to end of the list.
                
            Output: Returns a left rotated array.
            On Failure: Raise Exception

            Written By: Vishwaraj Kolekar
            Version: 1.0
            Revisions: None
        """
        logging.info("this is my rotate array function, my array is: %s and rotate by parameter is: %s", self.A, d)
        try:
            temp = []
            n=len(self.A)
            for i in range(d,n):
                temp.append(self.A[i])
            i = 0
            for i in range (0,d):
                temp.append(self.A[i])
            return temp
            logging.info("executed sucessfully")
            
        except Exception as e :
            print("this is my error msg " , e )
            logging.error("Error has happended")
            logging.exception("Exception occured" + str(e))
        

    def split_array(self, p):
        """
            Method Name: split array
            Description: This array splits array from a specified position, and move the first part of array and add to the end.
            Output: Splitted rearranged array.
            On Failure: Raise Exception

            Written By: Vishwaraj Kolekar
            Version: 1.0
            Revisions: None            
        """
        logging.info("this is my find largest number function, my array is: %s and split position is: %s", self.A , p)
        try:
            temp = []
            finalarray = []
            n = len(self.A)
            for i in range(p):
                temp.append(self.A[i])
            i = p
            for i in range(p,n):
                finalarray.append(self.A[i])
            i = 0
            for i in temp:
                finalarray.append(i)
            return finalarray
            
            logging.info("executed successfully")
        
        except Exception as e :
            print("this is my error msg " , e )
            logging.error("Error has happended")
            logging.exception("Exception occured" + str(e))

    def ismonotone(self):
        """
            Method Name: ismonotome
            Description: This method checks if given array is monotome or not
            Output: Gives Boolean output (True or False).
            On Failure: Raise Exception

            Written By: Vishwaraj Kolekar
            Version: 1.0
            Revisions: None            
        """
        logging.info("this is my monotone function, my array is: %s ", self.A)
        try:
            return (all(self.A[i] <= self.A[i + 1] for i in range(len(self.A) - 1)) or
                    all(self.A[i] >= self.A[i + 1] for i in range(len(self.A) - 1)))
            
            logging.info("executed sucessfully")
            
        except Exception as e :
            print("this is my error msg " , e )
            logging.error("Error has happended")
            logging.exception("Exception occured" + str(e))
```
