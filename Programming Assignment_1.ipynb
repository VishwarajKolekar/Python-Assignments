3. Write a Python program to find the area of a triangle?

```python
def triangle_area(h,b):
    area = 0.5*h*b
    return area

def equi_Triangle_area(a):
    area = (3**0.5)/4*(a**2)
    return area
```

```python
equi_Triangle_area(5)
```

```python
triangle_area(4,3)
```

4. Write a Python program to swap two variables?

```python
def swap(a,b):
    t = a
    a = b
    b = t
    return a,b
```

```python
swap(6,3)
```

5. Write a Python program to generate a random number?

```python
import random
def random_no(a,b):
    x = random.randint(a,b)
    return x
```

```python
random_no(1,100)
```
