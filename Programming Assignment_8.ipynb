### 1. Write a Python Program to Add Two Matrices?

```python
import logging
logging.basicConfig(filename = "assignment8.log", level = logging.DEBUG, format = '%(asctime)s %(levelname)s %(message)s')

class assignment8():
    """
        This class is used for assignment 8 operations.
        
        Written By: Vishwaraj Kolekar
        Version: 1.0
        Revisions: None
        
    """
    def __init__(self,a, b):
        self.a = a
        self.b = b
        
    def addmat(self):
        """
                Method Name: sum of array
                Description: This method adds 2 matrices.
                Output: Gives matrix as sum of two input matrices .
                On Failure: Raise Exception

                Written By: Vishwaraj Kolekar
                Version: 1.0
                Revisions: None

        """
        logging.info("this is my function to add two matrices")
        try:
            result = [[self.a[i][j] + self.b[i][j]  for j in range(len(self.a[0]))] for i in range(len(self.a))]
            for r in result:
                print(r)
            logging.info("executed successfully")
            
        except Exception as e :
            print("this is my error msg : " , e )
            logging.error("Error has happended")
            logging.exception("Exception occured" + str(e))

    def mulmat(self):
        """
            Method Name: mulmat
            Description: This method multiplies 2 matrices.
            Output: Gives matrix as product of two input matrices.
            On Failure: Raise Exception

            Written By: Vishwaraj Kolekar
            Version: 1.0
            Revisions: None
        """
        logging.info("this is my function to multiply two matrices")
        try:
            result = [[sum(i*j for i,j in zip(a_row,b_col)) for b_col in zip(*self.b)] for a_row in self.a]
            for r in result:
                print(r)
            logging.info("executed successfully")
            
        except Exception as e :
            print("this is my error msg " , e )
            logging.error("Error has happended")
            logging.exception("Exception occured" + str(e))
        

    def transmat(self):
        """
            Method Name: transmat
            Description: This method transposes a matrix.    
            Output: Gives a transposed matrix of input matrix.
            On Failure: Raise Exception

            Written By: Vishwaraj Kolekar
            Version: 1.0
            Revisions: None
        """
        logging.info("this is my function to transpose a matrix")
        try:
            result = [[self.a[j][i] for j in range(len(self.a))] for i in range(len(self.a[0]))]
            for r in result:
                print(r)
            logging.info("executed sucessfully")
            
        except Exception as e :
            print("this is my error msg " , e )
            logging.error("Error has happended")
            logging.exception("Exception occured" + str(e))
        

    def sort_alphabets(self):
        """
            Method Name: sort_alphabets
            Description: This method sorts words in string in alphabetic order.
            Output: words in alphabetic order.
            On Failure: Raise Exception

            Written By: Vishwaraj Kolekar
            Version: 1.0
            Revisions: None            
        """
        logging.info("this is my function to sort words in alphabetic order")
        try:
            s = input("Enter a string: ")
            words = [word.lower() for word in s.split()]
            words.sort()
            print("The sorted words are:")
            for word in words:
                print(word)
            
            logging.info("executed successfully")
        
        except Exception as e :
            print("this is my error msg " , e )
            logging.error("Error has happended")
            logging.exception("Exception occured" + str(e))

    def remove_punctuation(self):
        """
            Method Name: remove_punctuation
            Description: This method removes punctuations from a input string.
            Output: Gives a string with punctuations removed.
            On Failure: Raise Exception

            Written By: Vishwaraj Kolekar
            Version: 1.0
            Revisions: None            
        """
        logging.info("this is function to remove punctuations from a string")
        try:
            s = input("Enter a string with punctuations: ")
            punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
            final_string = ""
            for char in s:
                if char not in punctuations:
                    final_string = final_string + char
            print(final_string)
            
            logging.info("executed sucessfully")
            
        except Exception as e :
            print("this is my error msg " , e )
            logging.error("Error has happended")
            logging.exception("Exception occured" + str(e))
```
